````python
async def to_libreoffice(websocket: WebSocketServer):
    async for message in websocket:
        print(f"[SIDECAR]: {message}")
        lo_writer.write(message)
        await lo_writer.drain()
    
async def from_libreoffice(websocket: WebSocketServer):
    while True:
        data = await lo_reader.read(1)
        print(f"[LIBREOFFICE]: {data}")
        await websocket.send(data)

async def libreoffice(websocket: WebSocketServer):
    await asyncio.gather(
        to_libreoffice(websocket),
        from_libreoffice(websocket)
    )
````
